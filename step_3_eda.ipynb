# Step 3: EDA

```sql
SELECT TOP 10 *
FROM stock_data
```

```sql
-- How many companies are listed in this dataset?

SELECT COUNT(DISTINCT company) AS num_companies
FROM stock_data
```

```sql
-- What is the earliest date referecned in this dataset?

SELECT
    MIN([date]) AS earliest_date, 
    MAX([date]) AS latest_date
FROM stock_data
```

```sql
-- How much price fluctuation (at close) has each company experienced?

SELECT
    company, 
    ROUND(MAX([close]) - MIN([close]), 2) AS price_dif, 
    ROUND((MAX([close]) - MIN([close])) / MIN([close]) * 100, 2) as pct_change
FROM stock_data
GROUP BY company
```

```sql
-- How much volatility did each company experience? 

SELECT
    company, 
    STDEV([close]) AS price_volatility
FROM stock_data
GROUP BY company; 
```

```sql
-- What was the average closing stock price for each company? 

SELECT
    company, 
    AVG([close]) AS avg_close_price
FROM stock_data
GROUP BY company
```

```sql
-- What were the minimum and maximum volumes for each company? 

SELECT
    company, 
    MIN(volume) AS min_volume, 
    MAX(volume) AS max_volume
FROM stock_data
GROUP BY company
```

```sql
-- How does each company's opening vs. closing price compare (intraday swings)?

SELECT
    company,
    AVG([close] - [open]) AS avg_price_change_per_day
FROM stock_data
GROUP BY company;
```

```sql
-- What is each company's average closing price by month? 

SELECT
    company,
    [date],
    YEAR([date]) AS year,
    MONTH([date]) AS month,
    AVG([close]) AS avg_monthly_close
FROM stock_data
GROUP BY company, YEAR([date]), MONTH([date]), [date]
ORDER BY company, [date];
```

```sql
-- What was each company's top gain and loss day? 

SELECT 
    company,
    [date],
    (([close] - [open]) / [open]) * 100 AS pct_change_day
FROM stock_data
WHERE ([close] - [open]) / [open] = (
    SELECT MAX(([close] - [open]) / [open]) 
    FROM stock_data AS sd
    WHERE sd.company = stock_data.company
)
OR ([close] - [open]) / [open] = (
    SELECT MIN(([close] - [open]) / [open]) 
    FROM stock_data AS sd
    WHERE sd.company = stock_data.company
)
ORDER BY company, pct_change_day DESC;
```

```sql
-- What was each company's longest streak of positive / negative movement? 

WITH streaks AS (
    SELECT 
        company, 
        [date], 
        [close], 
        LEAD([close]) OVER (PARTITION BY company ORDER BY [date]) AS next_close,
        LAG([close]) OVER (PARTITION BY company ORDER BY [date]) AS prev_close,
        CASE 
            WHEN [close] > LAG([close]) OVER (PARTITION BY company ORDER BY [date]) THEN 1
            WHEN [close] < LAG([close]) OVER (PARTITION BY company ORDER BY [date]) THEN -1
            ELSE 0
        END AS streak_direction
    FROM stock_data
)
, streak_lengths AS (
    SELECT
        company,
        streak_direction,
        COUNT(*) AS streak_length
    FROM streaks
    WHERE streak_direction != 0
    GROUP BY company, streak_direction
)
SELECT
    company,
    MAX(CASE WHEN streak_direction = 1 THEN streak_length ELSE 0 END) AS longest_positive_streak,
    MAX(CASE WHEN streak_direction = -1 THEN streak_length ELSE 0 END) AS longest_negative_streak
FROM streak_lengths
GROUP BY company
ORDER BY company;
```

```sql
-- Mimic python's .describe() function on the 'close' column

WITH percentiles AS (
    SELECT 
        company,
        PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY [close]) OVER (PARTITION BY company) AS q1,
        PERCENTILE_CONT(0.50) WITHIN GROUP (ORDER BY [close]) OVER (PARTITION BY company) AS median,
        PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY [close]) OVER (PARTITION BY company) AS q3
    FROM stock_data
)
SELECT
    stock_data.company,
    ROUND(AVG([close]), 2) AS mean,
    ROUND(MIN([close]), 2) AS min,
    ROUND(MAX([close]), 2) AS max,
    ROUND(STDEV([close]), 2) AS stddev,  
    ROUND(VAR([close]), 2) AS variance, 
    q1,
    median,
    q3
FROM stock_data
LEFT JOIN percentiles ON stock_data.company = percentiles.company
GROUP BY stock_data.company, q1, median, q3;
```
